10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> Difference between map(), reduce(), and filter() in Python

map()

Applies a function to each element of an iterable and returns a new iterable with the results.

    Example:

nums = [1, 2, 3, 4]

squares = list(map(lambda x: x*x, nums))

print(squares)   # [1, 4, 9, 16]


filter()

Filters elements of an iterable based on a condition (True/False) and returns only those that satisfy it.

    Example:

nums = [1, 2, 3, 4, 5]

evens = list(filter(lambda x: x % 2 == 0, nums))

print(evens)   # [2, 4]


reduce() (from functools module)

Repeatedly applies a function to the elements of an iterable, reducing it to a single value.

    Example:

from functools import reduce

nums = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, nums)

print(product)   # 24

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

print(sum_even([1, 2, 3, 4, 5, 6]))
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number

```python
def square_list(nums):
    result = []
    for n in nums:
        result.append(n * n)
    return result

print(square_list([1, 2, 3, 4]))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = []
for num in range(1, 201):
    if is_prime(num):
        primes.append(num)

print(primes)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.i = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        if self.i == 0:
            self.i += 1
            return 0
        elif self.i == 1:
            self.i += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.i += 1
            return self.b

for num in Fibonacci(10):
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

for val in powers_of_two(5):
    print(val, end=" ")
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius = [0, 20, 30, 37, 100]
fahrenheit = list(map(lambda x: (x * 9/5) + 32, celsius))
print(fahrenheit)
```
